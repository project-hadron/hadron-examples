## Linear Regression L2 Regularisation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make the data
N = 50

# generate the data
X = np.linspace(0,10,N)
Y = 0.5*X + np.random.randn(N)

# make outliers
Y[-1] += 30
Y[-2] += 30
```

```python
# plot the data
plt.scatter(X, Y)
plt.show()
```

```python
# add bias term
X = np.vstack([np.ones(N), X]).T
```

### Residule Sum of Squares

$$ E = \sum_{n=1}^N \left( y_n - \hat y_n \right)^2 $$

```python
# plot the maximum likelihood solution
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
Yhat_ml = X.dot(w_ml)
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yhat_ml)
plt.show()
```

### L2 Regularization

$$ L2 = \sum_{n=1}^N \left(y_n -\hat{y_n} \right)^2 + \lambda \lVert w \rVert ^2 $$

```python
# L2 regularization
# probably don't need an L2 regularization this high in many problems
# everything in this example is exaggerated for visualization purposes
l2 = 1000.0

# add regularisation to What's the roomour slop (l2 * identity matrix + max likelyhood)
w_map = np.linalg.solve(l2*np.eye(2) + X.T.dot(X), X.T.dot(Y))

# calculate the Y estimate
Yhat_map = X.dot(w_map)
```

```python
# plot the regularized solution
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yhat_ml, label='maximum likelihood')
plt.plot(X[:,1], Yhat_map, label='map')
plt.legend()
plt.show()
```
